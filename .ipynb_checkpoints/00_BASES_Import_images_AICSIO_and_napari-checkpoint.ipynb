# Importing images using AICSimageIO and viewing them with napari

AICSimageIO is a module allowing to load several microscopy image formats like CZI, LIF, and OME-TIFF.
Napari is a python viewer based on Qt that helps displaying multidimensional data.

![Image of napari](images/illustrations/napari.png)

Let's start by importing the necessary libraries :

```python
from aicsimageio import AICSImage, imread
import napari
import matplotlib.pyplot as plt
```

## Using AICSimageIO

Here is how to use the basic AICS syntax to load an image and get some info :

```python
# Get an AICSImage object
img = AICSImage("images/confocal_convallaria.czi", B=0)
img.data;  # returns 6D STCZYX numpy array
print('dimensions: ' +img.dims)  # returns string "STCZYX"
print(f'shape: {img.shape}')  # returns tuple of dimension sizes in STCZYX order

# Get 6D STCZYX numpy array
data = imread("images/confocal_convallaria.czi")
```

**For more info** on how to use the module, read the documentation here : [Aicsimageio documentation at readthedocs](https://allencellmodeling.github.io/aicsimageio/index.html)

## Viewing data with Napari vs Matplolib

Here's how to use **napari *within* a Jupyter notebook** :

```python
with napari.gui_qt():
    viewer = napari.view_image(img.data[0,0,:,0,:,:])  # launches napari GUI and viewer
```

And here's how to view the same data with **matplolib** :

```python
fig = plt.figure(figsize=(30,10))
nb_channels = img.shape[2]
for x in range(0,nb_channels):
    fig.add_subplot(1,nb_channels+1, x+1)
    plt.imshow(img.data[0,0,x,0,:,:])
```
