# Importing images using AICSimageIO and viewing them with napari

## 1. Introduction

- **AICSimageIO** is a python module allowing to load several microscopy image formats like TIFF, CZI, LIF, and OME-TIFF.
- **Napari** is an interactive python image viewer based on Qt that helps displaying multidimensional data.

Here's a short preview of how it looks like with the example file :

![Image of napari](images/illustrations/napari-brain-sample.gif)

## 2. Using AICSimageIO

Here is how to use the basic AICSimageIO syntax to load an image and get some info :

To start we need to importing the necessary libraries :

```python
from aicsimageio import AICSImage, imread
import napari
import matplotlib.pyplot as plt
```

```python
# Get an AICSImage object

img_path = "images/first-instar-brain.tif"

img = AICSImage(img_path)

img.data  # returns 6D STCZYX numpy array
print("dimensions: " + img.dims)  # returns string "STCZYX"
print(f"shape: {img.shape}")  # returns tuple of dimension sizes in STCZYX order

# Get 6D STCZYX numpy array
data = imread(img_path)
img_to_view = img.get_image_data("SZYX", C=0, T=0)  # returns 4D SZYX numpy array
```

**For more info** on how to use the module, read the documentation here : [Aicsimageio documentation at readthedocs](https://allencellmodeling.github.io/aicsimageio/index.html)

## 3. Viewing data with Napari vs Matplolib

### 3.1 The Napari viewer

Here's how to use **napari *within* a Jupyter notebook** :

```python
with napari.gui_qt():
    viewer = napari.view_image(img_to_view)  # launches napari GUI and viewer
```

### 3.2 Viewing images with matplotlib

And here's how to view the same data with **matplolib**. You will notice we have to go through a loop to display the three channels and we don't have an easy access to slices.

```python
fig = plt.figure(figsize=(30, 10))
nb_channels = img_to_view.shape[0]
for x in range(0, nb_channels):
    fig.add_subplot(1, nb_channels + 1, x + 1)
    plt.imshow(img_to_view[x, 0, :, :])
```

### 3.3 Viewing images interactively with matplolib and ipywidgets

We can put the previous code in a function and use a slider with **ipywidgets** to display and interactive view of the dataset.

```python
from ipywidgets import interact, IntSlider


def plot_img(slice):
    fig = plt.figure(figsize=(30, 10))
    nb_channels = img_to_view.shape[0]
    for x in range(0, nb_channels):
        fig.add_subplot(1, nb_channels + 1, x + 1)
        plt.imshow(img_to_view[x, slice, :, :])


interact(
    plot_img,
    slice=IntSlider(
        min=0, max=img_to_view.shape[1], step=1, value=0, continuous_update=False
    ),
)
```
