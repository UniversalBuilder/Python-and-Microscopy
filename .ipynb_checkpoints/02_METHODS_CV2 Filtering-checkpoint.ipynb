# Denoising using OpenCV

Here are multiple denoising filters showed using CV2 (in option) and matplotlib.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("images/embryos_noisy.tif", 0)

# cv2.imshow("original", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
kernel = np.ones((3,3), np.float32)/9
filt_2D = cv2.filter2D(img, -1, kernel)

# cv2.imshow("2D custom filter", filt_2D)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
blur = cv2.blur(img, (3,3))

# cv2.imshow("Blur", blur)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
gaussian_blur = cv2.GaussianBlur(img, (3,3), 0)

# cv2.imshow("Gaussian blur", gaussian_blur)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
median_blur = cv2.medianBlur(img, 3)

# cv2.imshow("Median filter", median_blur)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
bilateral_blur = cv2.bilateralFilter(img, 9, 75, 75)

# cv2.imshow("Bilateral filter", bilateral_blur)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(img, 50, 150)

# cv2.imshow("Canny Edges", edges)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
fig = plt.figure(figsize=(20,15))

fig.add_subplot(3,3,1)
plt.imshow(img, cmap='jet')
plt.title('Original')

fig.add_subplot(3,3,2)
plt.imshow(filt_2D, cmap='jet')
plt.title('CV2 Filter 2D')

fig.add_subplot(3,3,3)
plt.imshow(blur, cmap='jet')
plt.title('CV2 Blur')

fig.add_subplot(3,3,4)
plt.imshow(gaussian_blur, cmap='jet')
plt.title('CV2 Gaussian_blur')

fig.add_subplot(3,3,5)
plt.imshow(median_blur, cmap='jet')
plt.title('CV2 Median_blur')

fig.add_subplot(3,3,6)
plt.imshow(bilateral_blur, cmap='jet')
plt.title('CV2 Bilateral_blur')

fig.add_subplot(3,3,7)
plt.imshow(edges, cmap='gray')
plt.title('CV2 Canny edges')
```
